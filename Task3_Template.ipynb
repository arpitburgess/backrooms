## Data Preprocessing

The sensors return data (sensor readings) for specific moments in time. Each row represents a different moment. However, detecting what specific movement a user did requires more than just the sensor readings of a single instant. It requires multiple readings over a period of time. We will divide the data using "sliding windows". The default number I have given for the window is 10. This means 10 rows will be brought together into a single row (with overlap so think 1 to 10, 2 to 11, 3 to 13, etc). However, for each feature we'll want to calculate different metrics that combine the features in different ways. For example, we can find the average value for the feature foot_Accel_X in the window. Or the standard deviation. Or the max and min. You get the idea. You'll want to do this for each of the original features (i.e. foot_Accel_X	foot_Accel_Y	foot_Accel_Z	foot_Gyro_X	foot_Gyro_Y	foot_Gyro_Z	shank_Accel_X	shank_Accel_Y, etc).

```python
import pandas as pd

window_size = 100
df = pd.DataFrame()

# Create the sliding windows for the different features
rolling_columns = []

for column in raw_df.columns:
    if column != "Header" and column != "labels":
        rolling_data = raw_df[column].rolling(window=window_size)

        rolling_columns.extend([
            rolling_data.mean()[window_size-1::window_size].rename(f"{column}_mean"),
            rolling_data.std()[window_size-1::window_size].rename(f"{column}_std"),
            rolling_data.max()[window_size-1::window_size].rename(f"{column}_max"),
            rolling_data.min()[window_size-1::window_size].rename(f"{column}_min")
        ])

# Concatenate the rolling columns and the original DataFrame
df = pd.concat([df] + rolling_columns, axis=1)

# Process the "labels" column
rolling_labels = raw_df["labels"].rolling(window=window_size, min_periods=window_size).apply(lambda x: x.iloc[-1])[window_size-1::window_size]
df["labels"] = rolling_labels

# Drop NaN values
df = df.dropna()
```
